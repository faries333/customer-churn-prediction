```python
sns.countplot(x="Churn", data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
df.isnull().sum()
```

```python
categorical_cols = df.select_dtypes(include="object").columns
numerical_cols = df.select_dtypes(exclude="object").columns

categorical_cols, numerical_cols
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Churn", y="Customer service calls", data=df)
plt.title("Customer Service Calls vs Churn")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="International plan", hue="Churn", data=df)
plt.title("International Plan vs Churn")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Churn", y="Total day minutes", data=df)
plt.title("Total Day Minutes vs Churn")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="Churn", y="Total intl charge", data=df)
plt.title("Total International Charge vs Churn")
plt.show()
```

# **Key EDA Insights**


---


*  Customers with international plans exhibit a higher churn rate.
*  Higher total day minutes are associated with increased churn.
*  Customers with higher international charges tend to churn more.
*  Customer service call frequency is a strong indicator of churn risk.
*  Customers with international plans combined with high usage show elevated churn risk
